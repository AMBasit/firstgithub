# body mass index (BMI)
# input Weight
# Height
# BMI (Ask yourself your height)
# Weight
# Calculate BMI
# print BMI (My name is     , and my BMI is ------)

```python
inputName = input("Enter your cool name please... ")
inputWeight = int(input("Enter your weight in kg please... "))
inputHeight = float(input("Enter your height in meters please... "))
BMI = (inputWeight/inputHeight**2)
print("My name is ", inputName , "and my BMI is", BMI)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic = sns.load_dataset("titanic")
p1=sns.countplot(x='sex', data=titanic, hue='class')
p1.set_title("Plot for Counting")
plt.show()
```
